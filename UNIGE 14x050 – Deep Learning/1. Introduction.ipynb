```python
import torch, torchvision
cifar = torchvision.datasets.CIFAR10('./cifar10', train = True, download = True)
```

```python
x = torch.from_numpy(cifar.data).permute(0, 3, 1, 2).float() / 255
print(x.dtype, x.size(), x.min().item(), x.max().item())
```

```python
x = x[:48]
torchvision.utils.save_image(x, 'cifar-4x12.png', nrow = 12, pad_value = 1.0)
```

```python
x.transpose_(2, 3)
torchvision.utils.save_image(x, 'cifar-4x12-rotated.png', nrow = 12, pad_value = 1.0)
```

```python
x[0:, 1:3].fill_(0)
torchvision.utils.save_image(x, 'cifar-4x12-rotated-and-red.png', nrow = 12, pad_value = 1.0)
```

```python
torchvision.utils.save_image(x, 'cifar-4x12-rotated-and-red_0_padding.png', nrow = 12, pad_value = 0.0)
```
