```python
import torch
import torch.nn.functional as F

nb_symbols = 6
s = torch.tensor([0, 1, 2, 3, 2, 1, 0, 5, 0, 5, 0])
x = F.one_hot(s, num_classes = nb_symbols)
x
```

```python
from torch.nn.utils.rnn import pack_padded_sequence
pack_padded_sequence(torch.tensor([[[1.], [2.]],
                                   [[3.], [4.]],
                                   [[5.], [0.]]]),
                     torch.tensor([3, 2]))
```
